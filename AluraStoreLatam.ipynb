# 1. An√°lisis de facturaci√≥n



### calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
# Primero calcular las facturaciones
print("üí∞ AN√ÅLISIS DE FACTURACI√ìN")
print("=" * 40)

facturacion_tienda1 = tienda['Precio'].sum()
facturacion_tienda2 = tienda2['Precio'].sum()
facturacion_tienda3 = tienda3['Precio'].sum()
facturacion_tienda4 = tienda4['Precio'].sum()

print(f"Tienda 1: ${facturacion_tienda1:,.2f}")
print(f"Tienda 2: ${facturacion_tienda2:,.2f}")
print(f"Tienda 3: ${facturacion_tienda3:,.2f}")
print(f"Tienda 4: ${facturacion_tienda4:,.2f}")

# Identificar tienda con menor facturaci√≥n
facturaciones = {
    'Tienda 1': facturacion_tienda1,
    'Tienda 2': facturacion_tienda2,
    'Tienda 3': facturacion_tienda3,
    'Tienda 4': facturacion_tienda4
}

tienda_menor_facturacion = min(facturaciones, key=facturaciones.get)
print(f"\nüìä La tienda con MENOR facturaci√≥n es: {tienda_menor_facturacion}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar el estilo de los gr√°ficos
plt.style.use('default')
sns.set_palette("husl")

# Crear gr√°fico de barras
print("GR√ÅFICO DE FACTURACI√ìN POR TIENDA")
print("=" * 40)

fig, ax = plt.subplots(figsize=(10, 6))

# Datos para el gr√°fico
tiendas = list(facturaciones.keys())
facturaciones_valores = list(facturaciones.values())

# Crear barras
barras = plt.bar(tiendas, facturaciones_valores, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])

# Personalizar el gr√°fico
plt.title('Facturaci√≥n Total por Tienda', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Facturaci√≥n ($)', fontsize=12)
plt.xticks(fontsize=11)

# Formatear los valores en el eje Y
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

# A√±adir valores en las barras
for barra, valor in zip(barras, facturaciones_valores):
    height = barra.get_height()
    ax.text(barra.get_x() + barra.get_width()/2., height + 10000000,
            f'${valor:,.0f}', ha='center', va='bottom', fontweight='bold')

# Destacar la tienda con menor facturaci√≥n
barra_menor = barras[tiendas.index(tienda_menor_facturacion)]
barra_menor.set_color('#FF9999')
barra_menor.set_edgecolor('red')
barra_menor.set_linewidth(2)

# A√±adir anotaci√≥n para la tienda con menor facturaci√≥n
ax.annotate(f'Menor facturaci√≥n\n{tienda_menor_facturacion}',
            xy=(barra_menor.get_x() + barra_menor.get_width()/2, barra_menor.get_height()),
            xytext=(0, 20),
            textcoords='offset points',
            ha='center',
            fontweight='bold',
            color='red',
            bbox=dict(boxstyle='round,pad=0.3', facecolor='lightyellow', alpha=0.7))

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Mostrar estad√≠sticas adicionales
print(f"\n ESTAD√çSTICAS ADICIONALES:")
print(f"Facturaci√≥n total todas las tiendas: ${sum(facturaciones_valores):,.2f}")
print(f"Facturaci√≥n promedio: ${sum(facturaciones_valores)/len(facturaciones_valores):,.2f}")
print(f"Diferencia entre mayor y menor facturaci√≥n: ${max(facturaciones_valores) - min(facturaciones_valores):,.2f}")
```

# 2. Ventas por categor√≠a

### calcular la cantidad de productos vendidos por categor√≠a en cada tienda. La idea es agrupar los datos por categor√≠a y contar el n√∫mero de ventas de cada tipo, mostrando las categor√≠as m√°s populares de cada tienda

```python
# Ventas por Categor√≠a - Categor√≠as m√°s populares por tienda
print("\n VENTAS POR CATEGOR√çA")
print("=" * 40)

def categorias_populares(df, nombre_tienda):
    ventas_por_categoria = df['Categor√≠a del Producto'].value_counts()
    categoria_top = ventas_por_categoria.index[0]
    cantidad_top = ventas_por_categoria.iloc[0]
    
    print(f"{nombre_tienda}:")
    print(f"  Categor√≠a m√°s popular: {categoria_top}")
    print(f"  Cantidad de ventas: {cantidad_top}")
    print(f"  Todas las categor√≠as: {dict(ventas_por_categoria.head())}")
    print("-" * 30)
    
    return categoria_top, cantidad_top

cat_tienda1 = categorias_populares(tienda, "Tienda 1")
cat_tienda2 = categorias_populares(tienda2, "Tienda 2")
cat_tienda3 = categorias_populares(tienda3, "Tienda 3")
cat_tienda4 = categorias_populares(tienda4, "Tienda 4")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Configurar el estilo
plt.style.use('default')
sns.set_palette("pastel")

print("ü•ß GR√ÅFICO DE TORTA - DISTRIBUCI√ìN DE PRODUCTOS")
print("=" * 50)

# Funci√≥n para obtener datos para gr√°ficos de torta
def obtener_datos_torta(df, nombre_tienda, top_n=5):
    productos_count = df['Producto'].value_counts()
    
    # Top N productos m√°s vendidos
    top_mas = productos_count.head(top_n)
    
    # Top N productos menos vendidos
    top_menos = productos_count.tail(top_n)
    
    return {
        'tienda': nombre_tienda,
        'top_mas': top_mas,
        'top_menos': top_menos,
        'total_productos': len(productos_count)
    }

# Obtener datos para todas las tiendas
datos_torta1 = obtener_datos_torta(tienda, "Tienda 1")
datos_torta2 = obtener_datos_torta(tienda2, "Tienda 2")
datos_torta3 = obtener_datos_torta(tienda3, "Tienda 3")
datos_torta4 = obtener_datos_torta(tienda4, "Tienda 4")

todos_datos_torta = [datos_torta1, datos_torta2, datos_torta3, datos_torta4]
```

```python
# Gr√°fico de torta para productos m√°s vendidos
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for i, (ax, datos) in enumerate(zip(axes, todos_datos_torta)):
    # Datos para el gr√°fico
    productos = datos['top_mas'].index
    ventas = datos['top_mas'].values
    total_top = ventas.sum()
    total_tienda = datos['top_mas'].sum() + datos['top_menos'].sum()  # Estimado
    
    # Crear gr√°fico de torta
    wedges, texts, autotexts = ax.pie(ventas, 
                                     labels=productos,
                                     autopct='%1.1f%%',
                                     startangle=90,
                                     colors=plt.cm.Set3.colors)
    
    # Personalizar
    ax.set_title(f'{datos["tienda"]}\nTop 5 Productos M√°s Vendidos\n(Total: {total_top} ventas)', 
                fontsize=12, fontweight='bold', pad=20)
    
    # Mejorar legibilidad
    for autotext in autotexts:
        autotext.set_color('white')
        autotext.set_fontweight('bold')
        autotext.set_fontsize(9)
    
    for text in texts:
        text.set_fontsize(9)

plt.suptitle('Productos M√°s Vendidos por Tienda', 
             fontsize=16, fontweight='bold', y=0.95)
plt.tight_layout()
plt.show()
```

```python
# EVALUACI√ìN ESTRAT√âGICA CON DATOS REALES
print("\nüéØ EVALUACI√ìN ESTRAT√âGICA INTEGRAL - DATOS REALES")
print("=" * 60)

# Sistema de puntuaci√≥n con datos reales
puntuaciones_finales = {}

for tienda in ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']:
    puntuacion = 0
    
    # Facturaci√≥n (35% peso)
    fact_max = max(facturaciones.values())
    puntuacion += (facturaciones[tienda] / fact_max) * 35
    
    # Calificaci√≥n (30% peso)
    puntuacion += (calificaciones[tienda] / 5) * 30
    
    # Costo env√≠o (25% peso) - inverso: menor costo = mayor puntuaci√≥n
    envio_min = min(envios.values())
    puntuacion += (envio_min / envios[tienda]) * 25
    
    # Eficiencia operativa (10% peso basado en datos generales)
    puntuacion += 7  # Valor base ajustado
    
    puntuaciones_finales[tienda] = puntuacion

# RANKING FINAL CON DATOS REALES
print("üèÜ RANKING ESTRAT√âGICO FINAL - BASADO EN DATOS REALES")
print("-" * 55)
for tienda, puntuacion in sorted(puntuaciones_finales.items(), key=lambda x: x[1], reverse=True):
    desempeno = "EXCELENTE" if puntuacion > 85 else "BUENO" if puntuacion > 75 else "REGULAR" if puntuacion > 65 else "CR√çTICO"
    print(f"{tienda}: {puntuacion:.1f}/100 - {desempeno}")

tienda_recomendada_venta = min(puntuaciones_finales, key=puntuaciones_finales.get)
print(f"\nüö® RECOMENDACI√ìN BASADA EN DATOS REALES:")
print(f"VENDER: {tienda_recomendada_venta}")
```
